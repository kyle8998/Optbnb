```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import statsmodels.api as sm
```

```python
# Set dtypes for each column later.
data = pd.read_csv('listings.csv', na_values=[""], dtype=object)

# Reads the neighbourhood csv file
neighbourhoods = pd.read_csv('neighbourhoods.csv', na_values=[""], dtype=object, usecols=range(1,2))

#data

# Sets the price column to floats without the '$' for computations
data['price'] = (data['price'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))

# Set ratings to floats
data['review_scores_accuracy'] = (data['review_scores_accuracy'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
data['review_scores_cleanliness'] = (data['review_scores_cleanliness'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
data['review_scores_checkin'] = (data['review_scores_checkin'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
data['review_scores_communication'] = (data['review_scores_communication'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
data['review_scores_location'] = (data['review_scores_location'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
data['review_scores_value'] = (data['review_scores_value'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
data['review_scores_rating'] = (data['review_scores_rating'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))

neighbourhoods['neighbourhood']
```

```python
# Regression graph data

w = 1
h = len(neighbourhoods);
averages = [[0 for x in range(w)] for y in range(h)]

for i in range(len(neighbourhoods)):
    neighbourhood_listings1 = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i]) & (data['bedrooms'] == '1')]['price']
    neighbourhood_listings2 = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i]) & (data['bedrooms'] == '2')]['price']
    neighbourhood_listings3 = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i]) & (data['bedrooms'] == '3')]['price']
    neighbourhood_listings4 = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i]) & (data['bedrooms'] >= '4')]['price']

    avg_price1 = np.mean(neighbourhood_listings1, 0)
    avg_price2 = np.mean(neighbourhood_listings2, 0)
    avg_price3 = np.mean(neighbourhood_listings3, 0)
    avg_price4 = np.mean(neighbourhood_listings4, 0)

    averages[i][0] = neighbourhoods['neighbourhood'][i]
    averages[i].append(avg_price1)
    print(avg_price1, ",", end="")
    averages[i].append(avg_price2)
    averages[i].append(avg_price3)
    averages[i].append(avg_price4)
    
avg_df = pd.DataFrame(averages,columns=['Neighbourhood','1', '2', '3', '4+'])
avg_df
```

```python
# Room to bathroom ratio graph

w = 4
h = 4;
room_to_bath = [[0 for x in range(w)] for y in range(h)]

for room in range(1, 5):
    for bath in range(1, 5):
        ratio = data.loc[(data['bathrooms'] == str(float(bath))) & (data['bedrooms'] == str(room))]['price']

        
            
        avg_price = np.mean(ratio, 0)
        room_to_bath[bath-1][room-1] = avg_price
        if room == 2 and bath == 4:
            print(ratio)
        
rtb_df = pd.DataFrame(room_to_bath,columns=['1 Room', '2 Rooms', '3 Rooms', '4 Rooms'])
for idx in range(4):
    print(rtb_df['1 Room'][idx])
rtb_df
```

```python
# Radar Review Graph

w = 1
h = len(neighbourhoods);
ratings = [[0 for x in range(w)] for y in range(h)]

for i in range(len(neighbourhoods)):
    accuracy = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_accuracy']
    cleanliness= data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_cleanliness']
    checkin = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_checkin']
    communication = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_communication']
    location = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_location']
    value = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_value']
    overall = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i])]['review_scores_rating']
    
    avg_accuracy = np.mean(accuracy, 0)
    avg_cleanliness = np.mean(cleanliness, 0)
    avg_checkin = np.mean(checkin, 0)
    avg_communication = np.mean(communication, 0)
    avg_location = np.mean(location, 0)
    avg_value = np.mean(value, 0)
    avg_overall = np.mean(overall, 0)
  

    ratings[i][0] = neighbourhoods['neighbourhood'][i]
    ratings[i].append(avg_accuracy)
    ratings[i].append(avg_cleanliness)
    ratings[i].append(avg_checkin)
    ratings[i].append(avg_communication)
    ratings[i].append(avg_location)
    ratings[i].append(avg_value)
    ratings[i].append(avg_overall)
    print(ratings[i][7])
    
rating_df = pd.DataFrame(ratings,columns=['Neighbourhood', 'Accuracy', 'Cleanliness', 'Check-in', 'Communication', 'Location', 'Value', 'Overall Rating'])
rating_df

rating_df.to_csv('rating_data.csv')
```
