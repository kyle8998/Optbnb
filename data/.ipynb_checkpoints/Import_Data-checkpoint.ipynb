```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import statsmodels.api as sm
```

```python
# Set dtypes for each column later.
data = pd.read_csv('listings.csv', na_values=[""], dtype=object)

# Reads the neighbourhood csv file
neighbourhoods = pd.read_csv('neighbourhoods.csv', na_values=[""], dtype=object, usecols=range(1,2))

#data

# Sets the price column to floats without the '$' for computations
data['price'] = (data['price'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))

neighbourhoods['neighbourhood']
```

```python
# Compare price means in each neighbourhood
prices = (data['price'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))

#neighbourhoods = data['neighbourhood']
bedrooms = data['bedrooms']
```

```python
#PvN = np.c_[prices, neighbourhoods]
#PvN[1][0]

mission_district = data
data['id']

# Get prices for a specific neighbourhood
specific = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][0]) & (data['bedrooms'] == '1')]['price']
specific
len(neighbourhoods)

w = 1
h = len(neighbourhoods);
CoL = [[0 for x in range(w)] for y in range(h)]

for i in range(len(neighbourhoods)):
    neighbourhood_listings = data.loc[(data['neighbourhood_cleansed'] == neighbourhoods['neighbourhood'][i]) & (data['bedrooms'] == '1')]['price']
    avg_price = np.mean(neighbourhood_listings, 0)
    CoL[i][0] = neighbourhoods['neighbourhood'][i]
    CoL[i].append(avg_price)
    
CoL_df = pd.DataFrame(CoL,columns=['Neighbourhood','Average Price'])
```

```python
CoL_df.plot(kind='bar', title='Average Daily Price in San Francisco', figsize=(20,5))
CoL_df.set_xlabel('a')
```
